```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group', as_index=False).count()

# there are 5 neighborhood groups and manhattan shows up the mostm byt brooklyn is very close to it. The other 3
# occur substantially less often
```

```python
# Are private rooms the most popular in manhattan?
# air_bnb.groupby('room_type', as_index=False).count()

manhattan = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].reset_index(drop=True)
manhattan.groupby('room_type', as_index=False).count()


# entire home/apt are by far the most popular. They are twice as popular as private rooms and 27 times more popular than shared rooms
```

```python
# Which hosts are the busiest and based on their reviews?
busiest =air_bnb.groupby('host_id', as_index=False).count().sort_values('id',ascending=False).head()

busiest_merge =  air_bnb.merge(busiest, on = 'host_id', how = 'inner')
busiest_merge_drop = busiest_merge.drop_duplicates(subset = 'host_id').reset_index(drop = True)
busiest_merge_drop[['host_id','host_name_x','price_y']].sort_values('price_y',ascending=False)


# host_id:     219517861    107434423     30283594      137358866     16098958
# host_name:   Sonder (NYC) Blueground    Kara          Kazuya        Jeremy & Laura
# listings:    327          232           121           103           96

#most = air_bnb.groupby('host_id')['name'].count().nlargest()
#most

```

```python
#Which neighorhood group has the highest average price?
display(air_bnb.groupby('neighbourhood_group', as_index=False).mean()[['neighbourhood_group','price']])

bronx = air_bnb[air_bnb['neighbourhood_group'] == 'Bronx'].reset_index(drop=True)
brooklyn = air_bnb[air_bnb['neighbourhood_group'] == 'Brooklyn'].reset_index(drop=True)
queens = air_bnb[air_bnb['neighbourhood_group'] == 'Queens'].reset_index(drop=True)
staten_island = air_bnb[air_bnb['neighbourhood_group'] == 'Staten Island'].reset_index(drop=True)
manhattan = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].reset_index(drop=True)

display(bronx.describe()['price'])
display(brooklyn.describe()['price'])
display(queens.describe()['price'])
display(staten_island.describe()['price'])
display(manhattan.describe()['price'])


# manhattan has the highest average prive of $196.88. It also has the highest median price of $150.
# It does have the highest standard deviation as well, which means it has the highest variability in price.
# having at least one value of 10000, which is 50 times higher than the mean is concerning, however this is
# represented for by the median being almost $50 below the mean, showing the data is skewed right and 
# is therefore not normal 

# it should be noted that all the means are higher than their corrosponding medians, so all the data is skewed right
# (this means that there are large outliars pulling the mean higher than it would otherwise be if the data was 
# normally distributed)
```

```python
# Which neighbor hood group has the highest total price?
display(air_bnb.groupby('neighbourhood_group', as_index=False).sum()[['neighbourhood_group','price']])
display(air_bnb.groupby('neighbourhood_group', as_index=False).count()[['neighbourhood_group','id']])

# Manhattan has the larget total price, followed by Brooklyn. This is not surprising since they are the two most
# common listings by far, Manhattan at 21661 and Brooklyn at 20104. What is surprising is that Manhattan's total 
# price is 1.7 times higher than Brooklyn's, desprite them having a roughly equal number of listings 

```

```python
#Which top 5 hosts have the highest total price?
tot_price = air_bnb.groupby('host_id', as_index=False).sum()[['host_id','price']].sort_values('price',ascending=False).head()

tot_price_merge =  air_bnb.merge(tot_price, on = 'host_id', how = 'inner')
tot_price_merge_drop = tot_price_merge.drop_duplicates(subset = 'host_id').reset_index(drop = True)
tot_price_merge_drop[['host_id','host_name','price_y']].sort_values('price_y',ascending=False)

# The top two in terms of listings are also the top two in terms of total price, however, the third in total 
# listings is 5th in total price. 3rd and 4th in total price are not in the top five of total listings

# host_id:     219517861    107434423     30283594      137358866     16098958
# host_name:   Sonder (NYC) Blueground    Sally         Red Awning    Kara        
# price:       82795        70331         37097         35294         33581
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb_zero = air_bnb[['host_id','number_of_reviews','availability_365']]

def zero_avail(row):
    num = row['number_of_reviews']
    avail = row['availability_365']
    if num >= 100 and avail == 0:
        availible = 'no'
    else:
        availible = 'yes'
    return availible

air_bnb_zero['Availible'] = air_bnb_zero.apply(zero_avail, axis = 1)
air_bnb_zero[air_bnb_zero['Availible'] == 'no'].reset_index(drop=True)

# There are currently 162 listings that have no availbility and over 100 reviews
```

```python
# What host has the highest total of prices and where are they located?
top_5 = air_bnb.groupby('host_id', as_index=False).sum().sort_values('price',ascending=False).head()

top_5_merge =  air_bnb.merge(top_5, on = 'host_id', how = 'inner')
top_5_merge_drop = top_5_merge.drop_duplicates(subset = 'host_id').reset_index(drop = True)
display(top_5_merge_drop[['host_id','host_name','neighbourhood_group','price_y']].sort_values('price_y',ascending=False))

# host_id:     219517861    107434423     30283594      137358866     16098958
# host_name:   Sonder (NYC) Blueground    Sally         Red Awning    Kara        
# price:       82795        70331         37097         35294         33581
#Neighbourhood Manhattan    Manhattan     Brokklyn      Manhattan     Manhattan
```

```python
# When did Danielle from Queens last receive a review?
air_bnb_D = air_bnb[air_bnb['host_name'] == 'Danielle'].reset_index(drop=True)
air_bnb_D_Queens = air_bnb_D[air_bnb_D['neighbourhood_group'] == 'Queens'].reset_index(drop=True)
air_bnb_D_Queens['last_review'] = air_bnb_D_Queens['last_review'].fillna('N/A')
air_bnb_D_Queens[['host_id','host_name','last_review','neighbourhood_group','name']]

# There are 4 different Danielle from Queens and 8 total listings. 
#host id             26432133           18051286         154256662      201647469
#last review data    2019-07-08         N/A              2018-01-02     2019-06-20
```
